#3) What does a wrong answer cost?

False positive: If a page is wrongly flagged as underperforming, the analyst wastes limited time working on a page that was already fine, and rewriting it risks introducing a small real decline to a page that was already performing well.

False negative: If a genuinely underperforming page is missed, its click-through problem goes unnoticed and continues to quietly cost the client lost traffic and potential revenue, for as long as it stays unflagged.
Between the two, a false negative is worse, a wasted afternoon is a bounded, one-time cost, while a missed real problem accumulates lost traffic and revenue over time. So I think the ranking should lean toward catching borderline cases rather than staying silent on them.

#4) Why does data or ML help at all?
A simple flat rule (Example: "CTR under 1%") isn't enough because expected CTR varies heavily by position, content type, and intent. A CTR that's normal at position 8 would be a serious problem at position 2. These signals interact in ways too tangled for one fixed threshold to capture fairly, which is exactly where a data-driven approach earns its place over a hardcoded rule.

# What I did and found:

To check whether Lane 4's problem actually exists in the data, I first tried a simple flat rule: pages with a top-10 average position and CTR under 1%. This flagged 39.4% of all pages, far too many to be believable or workable.
Digging deeper, I found that expected CTR varies heavily by position: average CTR drops from about 6.7% at position 1 to about 0.3% at position 10. A flat CTR cutoff unfairly penalizes pages at worse positions, where lower CTR is normal.
Instead, I calculated each page's CTR gap versus the average CTR for its own position tier. Using a moderate gap threshold (0.5 points below expected), 16.5% of all pages (4,955) show meaningful underperformance relative to their position; a real, sizeable opportunity, but one that requires ranking rather than blanket action, since not even 4,955 pages could be reviewed by hand in 7 weeks.

```python
import pandas as pd

df = pd.read_csv("flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv")
# df.head()
underperforming = df[
    (df["avg_position"] > 0) &
    (df["avg_position"] <= 10) &
    (df["ctr"] < 1.0)
]

print(f"Total pages: {len(df)}")
print(f"Underperforming (good position, bad CTR): {len(underperforming)}")
print(f"Percentage: {len(underperforming) / len(df) * 100:.1f}%")

top10 = df[(df["avg_position"] > 0) & (df["avg_position"] <= 10)]
print(top10.groupby(top10["avg_position"].round())["ctr"].mean())

expected_ctr = df.groupby(df["avg_position"].round())["ctr"].transform("mean")

df["ctr_gap"] = df["ctr"] - expected_ctr

top10_gap = df[(df["avg_position"] > 0) & (df["avg_position"] <= 10) & (df["ctr_gap"] < -0.5)]
print(f"Underperforming relative to position: {len(top10_gap)} ({len(top10_gap)/len(df)*100:.1f}%)")
```
