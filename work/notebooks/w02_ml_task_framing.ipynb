## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

#Answer
This is a scoring task, as I want to produce a continuous number for each page representing how much it underperforms its expected CTR for its position, not sort pages into fixed categories.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

#Answer
Target: 'trend_pct'. A continuous number already in the data, representing how much a page's performance has moved (example: -41.4 means a 41.4% decline).

Is it observed or defined? Observed, it's not something I calculated or invented; it already exists in the dataset as a real historical measurement.

What I'd predict it from: safe, non-circular features like search_volume, competition, content_type, main_intent, and word_count. Explicitly not ctr, avg_position, or anything derived from them, since those would leak the answer.

*Additional*

limitation: This starter dataset gives one snapshot per page, so this is associational, finding which content characteristics tend to go along with declining or improving trends, not true future forecasting. A real "predict what happens next" model would need the warehouse's daily time-series data (multiple rows per page over time), which isn't available in this starter dataset.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

#Answer
Metric: Mean Absolute Error (MAE) between predicted and actual trend_pct.
What "good" means: My model's MAE must be meaningfully lower than the MAE of a naive baseline that always guesses the dataset's average trend_pct for every page. If my model can't beat that simple baseline, it hasn't learned anything useful.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

#Answer
One row = one page (content_id), described by its features (search volume, competition, content type, main intent, word count) alongside the target, trend_pct. CTR, position, and other trend-derived columns are excluded from the feature set to avoid leaking the answer.



```python
import pandas as pd

df = pd.read_csv("flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv")

cols = ["content_id", "search_volume", "competition", "content_type",
        "main_intent", "word_count", "trend_pct"]

unit_df = df[cols]
print(f"Shape: {unit_df.shape}")
unit_df.head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

#Answer
A fixed rule can't easily capture this pattern because decline/growth likely depends on combinations of signals. For example, high competition might only hurt low-word-count pages, or search volume might matter across content types. These interactions are hard to hand-write as if-statements, but a model can learn them from data. This mirrors what I found in Week 1: CTR expectations varied by position in a way no single flat rule could fairly capture. The same kind of "many tangled signals" pattern applies here.
